# Facial Recognition using OpenCV

```python
%matplotlib inline
```

## Load dependencies

```python
import os
import cv2
import matplotlib.pyplot as plt
from scipy.misc import imread

# Mandatory: we need OpenCV3 library even 
# if we import `opencv2`
assert cv2.__version__ == '3.1.0'
```

```python
# Path to the pre-trained cascade classifier
face_cascade = os.path.join('..', 'data', 'classifiers', 'haarcascade_frontalface_default.xml')
# Path to the image
img_path = os.path.join('..', 'data', 'shots', 'Johnlieberman-SpecchioItalianRestaurantMiamiBeachSurfsideReviewByJohn753', '0343.jpg')
```

## Load external ressources

```python
# Load the cascade classifier
cascade = cv2.CascadeClassifier(face_cascade)
# Read image
img = imread(img_path)
# Preview the image we loaded
plt.imshow(img)
```

## Classify using the pre-trained cascade classifier

```python
# Convert img to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces using the pre-trained cascade classifier
faces = cascade.detectMultiScale(
    gray, scaleFactor=1.3, minNeighbors=5)
```

## Show faces founded

```python
# Draw a rectangle around the faces
for (x, y, w, h) in faces:
    res_img = cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

face_text = '{0} face'.format(len(faces)) if len(
            faces) == 1 else '{0} faces'.format(len(faces))

# Add the number of faces found
cv2.putText(res_img, face_text, (10, 30),
            cv2.FONT_HERSHEY_SIMPLEX, 1, 255)

# Finally, display the result
plt.imshow(res_img)
```
