# Exploratory analysis

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('ggplot')

%matplotlib inline
```

```python
attributes = pd.read_csv('../features/metadata/attributes.csv')
attributes['tag'] = pd.read_csv('../features/tags.csv')['tag_name']
```

```python
attributes.groupby('tag').boxplot('duration')
```

```python
tf_keywords = pd.read_csv('../features/metadata/tf_keywords.csv', index_col=0)
tf_keywords['tag'] = pd.read_csv('../features/tags.csv', index_col=0)['tag_name']
```

```python
top_n = {
    tag: group.drop('tag', axis=1).sum(axis=0).nlargest(20).keys().tolist()
    for tag, group in tf_keywords.groupby('tag')
}

pd.DataFrame(top_n)
```
