# KNN on color histograms

```python
def histogram_intersection(h1, h2):
    return sum(((min(v1, v2) for v1, v2 in zip(h1, h2))))
```

```python
import json

import pandas as pd


with open('../features/shots/hist_data.json', 'r') as in_file:
    histograms = json.load(in_file)

# Normalisation de tous les histogrammes
for img, histogram in histograms.items():
    histograms[img] = [bar / sum(histogram) for bar in histogram]
    
tags_df = pd.read_csv('../features/tags.csv')
```

```python
def predict_tag(img):
    # Détection des 9 images les plus proches
    neighbours = sorted(
        histograms.items(),
        key=lambda x: histogram_intersection(x[1], histograms[img])
    )[-10:]
    similar_pics = [neighbour[0] for neighbour in reversed(neighbours)]
    # Récupération les tags associés aux images les plus proches
    tags = [
        tags_df[tags_df['videofile'] == pic.split('/')[0]].tag_name.item()
        for pic in similar_pics
    ]
    # On retourne la valeur revenant le plus de fois
    return max(tags)
    
# Boucle donnant le df final
predictions = []
for i, img in enumerate(histograms.keys()):
    row = [str(img).split('/')[0], str(img).split('/')[1], str(predict_tag(str(img)))]
    predictions.append(row)
    if i % 20 == 0:
        print('{}/{}'.format(i, len(img))
```
