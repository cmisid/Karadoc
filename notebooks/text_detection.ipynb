## Tesseract OCR

We use the Python wrapper of __tesseract-ocr__ library. `pytesseract` contains a very usefull method called  `image_to_string` that takes in input a file handler.

```python
import pytesseract
from PIL import Image

text = pytesseract.image_to_string(Image.open(img_path))
print(text)
```
